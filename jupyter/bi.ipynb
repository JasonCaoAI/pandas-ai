```python
import pandas as pd
from pandasai.llm import AzureOpenAI
from pandasai import Agent
from dotenv import load_dotenv
import os
from IPython.display import display, HTML

load_dotenv()

os.unsetenv("ALL_PROXY")
os.unsetenv("http_proxy")
os.unsetenv("https_proxy")
os.environ["no_proxy"] = "localhost,127.0.0.1,http://chatgpt.tiny-test.wke-office.test.wacai.info"

llm = AzureOpenAI(
        api_token=os.getenv("API_KEY"),
        azure_endpoint=os.getenv("AZURE_ENDPOINT"),
        api_version=os.getenv("API_VERSION"),
        deployment_name=os.getenv("DEPLOYMENT_NAME"),
        temperature=0.0,
    )

cases = pd.read_excel("data/案件总表.xlsx")
goal = pd.read_excel("data/目标.xlsx")
agent = Agent([cases, goal], config={"llm": llm, "verbose": True, "save_logs": True})
result = agent.chat("将案件总表中的回收目标进行赋值，赋值的逻辑是：1. 根据案件总表中的案件类型和委托逾期等级字段去匹配 目标 表格中的类型和委托帐龄字段。2. 案件类型匹配逻辑为：“留案”匹配“留案”，其它匹配“当月新增”。3. 委托逾期等级匹配逻辑为需要将 逾期月份 被包含在 目标表格 的 委托帐龄字段的区间内：如M43匹配M36+，M11匹配M7-M12，M2匹配M2，满足这三种情况。将匹配目标中的 回收率_目标*案件总表中的考核分母 作为案件总表中的回收目标字段值。")
display(result)

# Display code with syntax highlighting
display(HTML(f"""
<pre style="background-color: #f6f8fa; padding: 10px; border-radius: 5px;">
<code>{agent.last_code_executed}</code>
</pre>
"""))
```

```python
cases = pd.read_excel("data/案件总表.xlsx")
goal = pd.read_excel("data/目标.xlsx")

dfs = [cases, goal]
案件总表 = dfs[0]
目标表格 = dfs[1]


def is_month_in_range(month_str, range_str):
    month = int(month_str[1:])
    if '+' in range_str:
        lower_bound = int(range_str[1:-1])
        return month >= lower_bound
    elif '-' in range_str:
        lower_str, upper_str = range_str.split('-')
        lower_bound, upper_bound = int(lower_str[1:]), int(upper_str[1:])
        return lower_bound <= month <= upper_bound
    else:
        return month == int(range_str[1:])


案件总表['回收目标'] = 0
for index, row in 案件总表.iterrows():
    案件类型 = row['案件类型']
    委案逾期等级 = row['委案逾期等级']
    考核分母 = row['考核分母']
    if 案件类型 == '留案':
        matching_type = '留案'
    else:
        matching_type = '当月新增'
    for _, target_row in 目标表格.iterrows():
        if target_row['类型'] == matching_type and is_month_in_range(委案逾期等级, target_row['委案账龄']):
            回收率_目标 = target_row['回收率_目标']
            案件总表.at[index, '回收目标'] = 回收率_目标 * 考核分母
            break
result = {'type': 'dataframe', 'value': 案件总表}

# Write the dataframe to CSV
result['value'].to_csv('案件总表_with_targets.csv', index=False, encoding='utf-8-sig')

result
```
